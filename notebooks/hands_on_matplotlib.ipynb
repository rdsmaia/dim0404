# Passos iniciais ao uso do matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definição de funções
def plot_xy(x, y, title):
    fig, ax = plt.subplots(1,figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    ax.set_xlabel('Valor de x')
    ax.set_ylabel('valor de y')
    plt.show()
```

```python
# função 1
x1 = np.arange(-50,51,0.1)
y1 = np.cbrt(x1) + x1**3 + 7
plot_xy(x1,y1, title='f(x)=x**(1/3)+x**3+7')
```

```python
# função 2
x2 = np.arange(-50,51,0.1)
y2 = np.abs(x2)
plot_xy(x2,y2, title='f(x)=|x|')
```

```python
# função 3
x3 = np.arange(-50,51,0.1)
y3 = np.sin(x3)
plot_xy(x3,y3, title='f(x)=sin(x)')
```

```python
# função 4
x4 = np.arange(0.1,10,0.1)
y4 = np.log(x4)
plot_xy(x4,y4, title='f(x)=ln(x)')
```

```python
# função 5
x5 = np.arange(-1,1,0.1)
y5 = np.exp(x5)
plot_xy(x5,y5, title='f(x)=exp(x)')
```
